```python
df = pd.read_csv('../Datasets/breast cancer.csv')
```

```python
df.drop(columns='Unnamed: 32', inplace=True)
```

```python
# Get the number of M and B in the dataset:
df.diagnosis.value_counts()
```

### Label Encoding:

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.diagnosis = le.fit_transform(df.diagnosis.values)
```

```python
df.head()
```

* Note the encoding

### Slpitting The Dataset & Feature Scaling:

```python
# Slpitting The Dataset Into Dependant/Independant Datasets:
x = df.iloc[:,2:]
y = df.diagnosis.to_frame()
```

```python
y
```

```python
x
```

```python
# Split to 75%/25% trin/test:
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

x_scaler = StandardScaler()
x_train = x_scaler.fit_transform(x_train)
x_test = x_scaler.fit_transform(x_test)
```

### Building The Logistic Regression Model:

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
classifier.fit(x_train,y_train.values.ravel())
```

```python
# Making predictions:
predictions = classifier.predict(x_test)
```

### Performance Evaluation:

```python
# Plot Confusion Matrix:
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predictions)
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
