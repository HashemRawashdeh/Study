```python
import os
import shutil
import random
from shutil import copyfile
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

# Splitting The Data Into Training/Validation:
---

```python
def split_data(source_dir, training_dir, validation_dir, split_size):
    # Shuffling the data:
    shuffled_source = random.sample(os.listdir(source_dir), len(os.listdir(source_dir)))
    
    # Find total number of files in training dir
    training_number = int(len(shuffled_source) * split_size)
    
    i=0
    target = training_dir
    for item in shuffled_source:
        item_dir = os.path.join(source_dir, item)
        if os.path.getsize(item_dir) == 0:
            print(f'{item} is of size  0, so ignoring')
        else:
            copyfile(item_dir, os.path.join(target, item))
            i += 1
        if i == training_number:
            target = validation_dir
```

```python
source_dir_dogs = '../../../../../Downloads/PetImages/Dog'
source_dir_cats = '../../../../../Downloads/PetImages/Cat'
training_dir_dogs = '../Datasets/cats-vs-dogs/Training/Dogs' 
training_dir_cats = '../Datasets/cats-vs-dogs/Training/Cats'
validation_dir_dogs = '../Datasets/cats-vs-dogs/Validation/Dogs' 
validation_dir_cats = '../Datasets/cats-vs-dogs/Validation/Cats'

split_data(source_dir_dogs, training_dir_dogs, validation_dir_dogs, 0.9)
split_data(source_dir_cats, training_dir_cats, validation_dir_cats, 0.9)
```

# Regular Approach:
---

```python
training_generator = ImageDataGenerator(rescale=1/255)

training_generator = training_generator.flow_from_directory('../Datasets/cats-vs-dogs/Training',
                                                            batch_size=45,
                                                            target_size=(150,150),
                                                            class_mode='binary')

validation_generator = ImageDataGenerator(rescale=1/255)
validation_generator = validation_generator.flow_from_directory('../Datasets/cats-vs-dogs/Validation',
                                                                batch_size=5,
                                                                target_size=(150,150),
                                                                class_mode='binary')
```

```python
model = tf.keras.models.Sequential([ 
                                    tf.keras.layers.Conv2D(16, 3, input_shape=(150,150,3)),
                                    tf.keras.layers.MaxPooling2D(),
                                    tf.keras.layers.Conv2D(32, 3),
                                    tf.keras.layers.MaxPooling2D(),
                                    tf.keras.layers.Conv2D(64, 3),
                                    tf.keras.layers.MaxPooling2D(),
                                    tf.keras.layers.Conv2D(128, 3),
                                    tf.keras.layers.MaxPooling2D(),
                                    tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(512, activation='relu'),
                                    tf.keras.layers.Dense(1, activation='sigmoid')
                                    ])
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics='accuracy')
```

```python
history = model.fit(training_generator,
                    epochs=15,
                    validation_data=validation_generator)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
epochs = range(len(acc))
plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
```

# Transfer Learning:
----
Here we will use the InceptionV3 network.

```python
# Download the pre-trained weights. No top means it excludes the fully connected layer it uses for classification.
!wget --no-check-certificate \
    https://storage.googleapis.com/mledu-datasets/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5 \
    -O inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5
```

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
```

```python
# Set the weights file you downloaded into a variable:
local_weights_file = 'inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5'
```

```python
pretrained_model = InceptionV3(input_shape=(150,150,3),
                               include_top=False,
                               weights=None)

pretrained_model.load_weights(local_weights_file)
for layer in pretrained_model.layers:
    layer.trainable=False
```

```python
pretrained_model.summary()
```

```python
last_layer = pretrained_model.get_layer('mixed7')
last_output = last_layer.output
```

```python
x = tf.keras.layers.Flatten()(last_output)
x = tf.keras.layers.Dense(units=1024, activation='relu')(x)
x = tf.keras.layers.Dropout(0.2)(x)
x = tf.keras.layers.Dense(1,'sigmoid')(x)

model = tf.keras.Model(pretrained_model.input, x)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics='acc')
```

```python
history = model.fit(training_generator,
                    epochs=10,
                    validation_data=validation_generator)
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
epochs = range(len(acc))
plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
```
