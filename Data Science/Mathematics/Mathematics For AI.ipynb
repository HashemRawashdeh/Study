### **Essence Of Linear Algebra Series**

**Videos:**
Here the videos are the first two videos of the essence of linear algebra series. A couple of eye opening viewpoints were coverd here, and I will mention them briefly, and do some examples to show the concepts. These ideas are:

1. You Can pick any arbitary basis vectors and write any vector in space using those basis vectors under the conditions that there are as many linearly independent basis vectors as the dimensions of the vector of question, and that they span the space.

2. You can look at a vector from three perspectives:
    1. The physics point of view: A thing in the physical space with a specific magnitude and direction.
    2. The computer science point of view: An ordered list of things.
    3. The mathematics point of view: Something that satisfy two properties: Vector addition, and scalar multiplication.


3. The basis set: some number of linearly independent vectors, and that number is the number of dimensions in that space. _Note that the choice of the basis set isn't unique._

4. The span: All the possible vectors generated from a linear combination of the basis set.

**_Note that the vectors in the basis set only need to satisfy linear independence, but choosing an orthonormal basis set simplifies the transformations alot._**

Here is an example: 

```python
from matplotlib import pyplot as plt

def draw_vector(org1x,org1y,vec, col,org2x,org2y,vec2,col2,org3x,org3y,vec3,col3,
                 org4x,org4y,vec4,col4,org5x,org5y,vec5,col5):
    '''This Function plots five two dimensional vectors by taking their components and their colors'''
    
    plt.figure()
    ax = plt.gca()
    
    ax.quiver(org1x,org1y,vec[0],vec[1], color=col, angles='xy', scale_units='xy',scale=1)
    ax.quiver(org2x,org2y,vec2[0],vec2[1], color=col2, angles='xy', scale_units='xy',scale=1)
    ax.quiver(org3x,org3y,vec3[0],vec3[1], color=col3, angles='xy', scale_units='xy',scale=1)
    ax.quiver(org4x,org4y,vec4[0],vec4[1], color=col4, angles='xy', scale_units='xy',scale=1)
    ax.quiver(org5x,org5y,vec5[0],vec5[1], color=col5, angles='xy', scale_units='xy',scale=1)
    
    ax.set_xlim([-5,5])
    ax.set_ylim([-5,5])
    plt.grid()
    
draw_vector(0,0,[1,0],'k',0,0,[0,1],'k', 0,0,[3,4], 'r',0,0,[2,1],'b',0,0,[-2,4],'b')  # Here you give the tail, each basis component and the color,
                                                                                       # respectively for each vector 
```

The two black vectors are an example of a basic basis set, which consists of two unit vectores, one horizontal and one vertical. The red vector is a vector in space that is represented as a scaled sum of the basis vectors. This vector can be represented as the sum of [3,0] and [0,4] which are 3 times the horizantal basis vector and four times the vertical one. These Two vectors come from projecting the red vector onto each basis vector, The resulting two vectors are shown in yellow below. I moved the vertical one three units to the right to show that their sum results in the red vector.

```python
draw_vector(0,0,[1,0],'k',0,0,[0,1],'k', 0,0,[3,4], 'r',0,0,[3,0],'y',3,0,[0,4],'y')
```

What the above showed is that the red vector can be represented by a linear sum of the basis vectors, i.e, the basis vectors span it. Since the choice of this basis vectors was claimed not to be unique, I will use the two blue vectors as a basis set, and represent the red vector using them.

```python
# Vector projecting the red vector onto the blue vectors:

red = np.array([3,4])
shorter_blue = np.array([2,1])
longer_blue = np.array([-2,4])

proj_short = (np.dot(red,shorter_blue) * shorter_blue) / (np.linalg.norm(shorter_blue)**2)
proj_long = (np.dot(red,longer_blue) * longer_blue) / (np.linalg.norm(longer_blue)**2)

draw_vector(0,0,[1,0],'k',0,0,[0,1],'k', 0,0,[3,4], 'r',0,0,[proj_short[0],proj_short[1]],'y',
            proj_short[0],proj_short[1],[proj_long[0],proj_long[1]],'y')
```

Note that we were also able to represent the red vector as a sum of the two blue basis vectors, i.e, the blue vectors span it as well and can be used as a bisis set. Note that the two blue vectors are orthogonal, and that's why the projection of the red vector onto them resulted in the red vector itself after addition. This is not the case for non-orthogonal vectors, and for that some other scaling and addition may be required, but that doesn't mean that we can't represent the red vector using them, it just means that it needs more steps. Try changing the shorter blue to [3,1] and notice the need for scaling. For a set of vectors to be a basis set, you only need to meet two conditions:
1. linear independence
2. Spans the space 

```python
shorter_blue = np.array([3,1])
proj_short = (np.dot(red,shorter_blue) * shorter_blue) / (np.linalg.norm(shorter_blue)**2)

draw_vector(0,0,[1,0],'k',0,0,[0,1],'k', 0,0,[3,4], 'r',0,0,[proj_short[0],proj_short[1]],'y',
            proj_short[0],proj_short[1],[proj_long[0],proj_long[1]],'y')
```

* Here is how projecting onto non-orthognal basis set and just adding them results to. Note how you need some scaling to represent the red vector.
* This is why orthogonal vectors are typically used as a basis set.

```python
colors = 'bgrcmykw'

def draw_vector():
    '''This function draws a randomely generated two dimensional vector in the two dimensional space
       resulting as a linear combination of two two dimensional vectors'''
    ax = plt.gca()
    x = np.array([np.random.randint(-40,40), np.random.randint(-40,40)], dtype=float)              # First two dimensional vector
    y = np.array([np.random.randint(-40,40), np.random.randint(-40,40)], dtype=float)              # Second two dimensional vector
    result = np.random.randint(-2,2) * x + np.random.randint(-2,2) * y                             # Their linear combination
    c = np.random.randint(0,8)
    ax.quiver(result[0], result[1], color=colors[c], angles = 'xy', scale_units='xy', scale=1)
```

```python
for i in range(500):
    draw_vector()
plt.xlim([-50,50])
plt.ylim([-50,50])
plt.grid()
```

* **_Note how the linear combination of the two basis vectors 500 times generated 500 vectors that start to span the entire space, and when we generate an infinite amount of vectors using linear combinations of the basis vectores, we will span the entire space._**
